```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

# Initialize SparkSession
spark = SparkSession.builder.appName("DataNormalizationSparkSQL").getOrCreate()

# Define the schema (Important: Make sure data types match your actual data)
schema = StructType([
    StructField("Question Response Text", StringType(), True),
    StructField("Question Response Type", StringType(), True),
    StructField("Subgroup Header", StringType(), True),
    StructField("Subgroup Element Text", StringType(), True),
    StructField("Base", IntegerType(), True),
    StructField("Abs Count", IntegerType(), True),
    StructField("Column Percent", DoubleType(), True),  # Or FloatType if preferred
    StructField("Annotation Left Header", StringType(), True),
    StructField("CustomField01_SurveyCategory", StringType(), True),
    StructField("CustomField02_ProjectName", StringType(), True),
    StructField("CustomField08_CommissioningMarket", StringType(), True),
    StructField("Annotation Center Header", StringType(), True)
])

# Load data from CSV
try:
    df = spark.read.csv("example_data.csv", header=True, schema=schema)
except Exception as e:
    print(f"Error reading CSV file: {e}")
    exit()

# Step 1: Data Extraction
#df.printSchema()
#df.show(10)

# Step 2: Data Transformation using Spark SQL

# Create temporary views for the DataFrames (required for Spark SQL)
df.createOrReplaceTempView("main_table")

# Create lookup tables (populate these fully!)
spark.sql("""
    CREATE OR REPLACE TEMPORARY VIEW brand_lookup AS
    SELECT 1 as brand_id, 'Brand A' as brand_name UNION ALL
    SELECT 2, 'Brand B' UNION ALL
    SELECT 4, 'Brand D'  -- Add all other brands except 'Brand C'
""")

spark.sql("""
    CREATE OR REPLACE TEMPORARY VIEW project_lookup AS
    SELECT 1 as project_id, 'Project Z' as project_name -- Add all other projects
""")

spark.sql("""
    CREATE OR REPLACE TEMPORARY VIEW market_lookup AS
    SELECT 1 as market_id, 'Portugal' as market_name -- Add all other markets
""")

# Perform transformations in a single SQL query
df_transformed = spark.sql("""
    SELECT
        `Question Response Type`, `Subgroup Header`, `Base`, `Abs Count`, `Column Percent`, `Annotation Left Header`, `CustomField01_SurveyCategory`, `Annotation Center Header`,
        COALESCE(bl.brand_id, CASE WHEN mt.`Question Response Text` = 'Brand C' THEN 3 END) as brand_id,  -- Dynamic Brand C
        pl.project_id,
        ml.market_id,
        TO_DATE(`Subgroup Element Text`, 'yyyyMMdd') as `SubgroupElementDate`  -- Convert to Date
    FROM main_table mt
    LEFT JOIN brand_lookup bl ON mt.`Question Response Text` = bl.brand_name
    LEFT JOIN project_lookup pl ON mt.`CustomField02_ProjectName` = pl.project_name
    LEFT JOIN market_lookup ml ON mt.`CustomField08_CommissioningMarket` = ml.market_name
""").cache() #cache the intermediate table

# Replace cache() with createOrReplaceTempView to register the DataFrame as a temporary view
df_transformed.createOrReplaceTempView("df_transformed")

# Step 2 (C): Rolling Average (using Spark SQL self join)

df_with_rolling_avg = spark.sql("""
    SELECT
        t1.*,
        AVG(t2.`Column Percent`) AS rolling_avg_4w
    FROM
        df_transformed t1
    INNER JOIN
        df_transformed t2 ON t1.brand_id = t2.brand_id AND t1.`Annotation Left Header` = t2.`Annotation Left Header` AND t2.SubgroupElementDate BETWEEN DATE_SUB(t1.SubgroupElementDate, 3) AND t1.SubgroupElementDate
    GROUP BY
        t1.`Question Response Type`, t1.`Subgroup Header`, t1.Base, t1.`Abs Count`, t1.`Column Percent`, t1.`Annotation Left Header`, t1.`CustomField01_SurveyCategory`, t1.`Annotation Center Header`, t1.brand_id, t1.project_id, t1.market_id, t1.SubgroupElementDate
    ORDER BY t1.brand_id, t1.`Annotation Left Header`, t1.SubgroupElementDate
""")


df_with_rolling_avg.printSchema()
df_with_rolling_avg.show()

spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

# Initialize SparkSession
spark = SparkSession.builder.appName("DataNormalizationSparkSQL").getOrCreate()

# Define the schema
schema = StructType([
    StructField("Question Response Text", StringType(), True),
    StructField("Question Response Type", StringType(), True),
    StructField("Subgroup Header", StringType(), True),
    StructField("Subgroup Element Text", StringType(), True),
    StructField("Base", IntegerType(), True),
    StructField("Abs Count", IntegerType(), True),
    StructField("Column Percent", DoubleType(), True),
    StructField("Annotation Left Header", StringType(), True),
    StructField("CustomField01_SurveyCategory", StringType(), True),
    StructField("CustomField02_ProjectName", StringType(), True),
    StructField("CustomField08_CommissioningMarket", StringType(), True),
    StructField("Annotation Center Header", StringType(), True)
])

# Load data from CSV
try:
    df = spark.read.csv("example_data.csv", header=True, schema=schema)
except Exception as e:
    print(f"Error reading CSV file: {e}")
    exit()

# Step 1: Data Extraction
#df.printSchema()
#df.show(100)

# Step 2: Data Transformation using Spark SQL
# Create temporary views
df.createOrReplaceTempView("main_table")

# Create lookup tables
spark.sql("""
    CREATE OR REPLACE TEMPORARY VIEW brand_lookup AS
    SELECT 1 as brand_id, 'Brand A' as brand_name UNION ALL
    SELECT 2, 'Brand B' as brand_name UNION ALL
    SELECT 4, 'Brand D' as brand_name  -- Add all other brands except 'Brand C'
""")

spark.sql("""
    CREATE OR REPLACE TEMPORARY VIEW project_lookup AS
    SELECT 1 as project_id, 'Project Z' as project_name -- Add all other projects
""")

spark.sql("""
    CREATE OR REPLACE TEMPORARY VIEW market_lookup AS
    SELECT 1 as market_id, 'Portugal' as market_name -- Add all other markets
""")

# Perform transformations and rolling average in a single SQL query
df_transformed = spark.sql("""
    WITH NormalizedData AS (
        SELECT
            `Question Response Type`, `Subgroup Header`, `Base`, `Abs Count`, `Column Percent`, `Annotation Left Header`, `CustomField01_SurveyCategory`, `Annotation Center Header`,
            COALESCE(bl.brand_id, CASE WHEN mt.`Question Response Text` = 'Brand C' THEN 3 END) as brand_id,  -- Dynamic Brand C
            pl.project_id,
            ml.market_id,
            TO_DATE(`Subgroup Element Text`, 'yyyyMMdd') as `date`  -- Convert to Date
        FROM main_table mt
        LEFT JOIN brand_lookup bl ON mt.`Question Response Text` = bl.brand_name
        LEFT JOIN project_lookup pl ON mt.`CustomField02_ProjectName` = pl.project_name
        LEFT JOIN market_lookup ml ON mt.`CustomField08_CommissioningMarket` = ml.market_name
    ),
    RollingAverage AS (
        SELECT
            t1.*,
            AVG(t2.`Column Percent`) AS rolling_avg_4w
        FROM
            NormalizedData t1
        INNER JOIN
            NormalizedData t2 ON t1.brand_id = t2.brand_id AND t1.`Annotation Left Header` = t2.`Annotation Left Header` AND t2.date BETWEEN DATE_SUB(t1.date, 3) AND t1.date
        GROUP BY
            t1.`Question Response Type`, t1.`Subgroup Header`, t1.Base, t1.`Abs Count`, t1.`Column Percent`, t1.`Annotation Left Header`, t1.`CustomField01_SurveyCategory`, t1.`Annotation Center Header`, t1.brand_id, t1.project_id, t1.market_id, t1.date
        ORDER BY t1.brand_id, t1.`Annotation Left Header`, t1.date
    )
    SELECT * FROM RollingAverage
""")


# If you have a dedicated location for unmanaged tables:
output_location = "/content/spark-warehouse/"
df_transformed.write.mode("overwrite").parquet(output_location)

#Store the transformed data into a new table with the following schema:
df_transformed.write.mode("overwrite").saveAsTable("final_table")

df_transformed.printSchema()
df_transformed.show(100)

spark.stop()
```
